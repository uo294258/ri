```python
import bm25s # To create indexes and search documents using BM25
import Stemmer  # For stemming terms
import json  # To load the JSON-formatted corpus
```

```python
!gdown 1q_8DhdlF9imN-ksNWAjrT-HpeO7wFR-4
!unzip -o trec-covid-RI.zip
```

```python
corpus_content = []  # Lista para almacenar los objetos JSON

with open("corpus.jsonl", "r", encoding="utf-8") as f:
    for line in f:  # Itera sobre cada línea del archivo
          json_object = json.loads(line)
          corpus_content.append(json_object)  # Agrega el objeto a la lista
# Ahora corpus_content es una lista que contiene todos los objetos JSON del archivo
print(corpus_content[0])  # Imprime el primer objeto JSON para verificar
```

```python
corpus_verbatim = list()
corpus_plaintext = list()
for entry in corpus_content:
    document = {"id": entry["_id"], "title": entry["title"], "text": entry["text"].lower()}
    corpus_verbatim.append(document)
    corpus_plaintext.append(entry["text"].lower())
```

```python
stemmer = Stemmer.Stemmer("english")

corpus_tokenized = bm25s.tokenize(corpus_plaintext, stopwords="en", stemmer=stemmer, show_progress=True)
```

```python
# We chose the lucene implementation, you can change it and try different options
bm25_flavor = "lucene"
idf_flavor = "lucene"

retriever = bm25s.BM25(corpus=corpus_verbatim, method=bm25_flavor, idf_method=idf_flavor)
retriever.index(corpus_tokenized, show_progress=True)
```

```python
retriever.save("trec-covid", corpus=corpus_verbatim)
```
